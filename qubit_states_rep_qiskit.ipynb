```python
from qiskit import QuantumCircuit, assemble, Aer
from qiskit.visualization import plot_histogram, plot_bloch_vector
from math import sqrt, pi
```

```python
# Creamos in circuito cuantico con un qubit
qc = QuantumCircuit(1)
# Definimos el estado inicial como |1> (matricialmente)
initial_state = [0,1]  
# Inicializamos el qubit 0
qc.initialize(initial_state, 0) 
# Dibujamos nuestro circuito
qc.draw("mpl")  
```

```python
# Podemos usar uno de los simuladores de Qiskit para ver el estado resultante de nuestro 
# qubit.

# Le indicamos a Qiskit como simular nuestro circuito
sim = Aer.get_backend('aer_simulator')  

# Para obtener los resultados de nuestro circuito, usamos "run" para ejecutar nuestro 
# circuito, dando el circuito y el backend como argumentos. Luego usamos ".result()" 
# para obtener el resultado de esto:

# Creamos el circuito cuantico
qc = QuantumCircuit(1) 
# Definimos el estado inicial como |1>
initial_state = [0,1]   
# Aplicamos la operacion de inicializacion sobre el qubir 0
qc.initialize(initial_state, 0) 
# Le indicamos al simulador que grabe el vector de estado
qc.save_statevector()   
# Creamos un Qobj del circuito para que se ejecute el simulador
qobj = assemble(qc)   
# Hacemos la simulacion y retornamos el resultado
result = sim.run(qobj).result() 
```

```python
# Podemos obtener el vector de estado final
out_state = result.get_statevector()
print(out_state) # Display the output state vector
```

```python
# Medimos nuestro qubit como lo haríamos en una computadora cuántica real y veamos el 
# resultado:
qc.measure_all()
qc.draw("mpl")
```

```python
# Conteos de los resultados 0 y 1 usando .get_counts():

qobj = assemble(qc)
result = sim.run(qobj).result()
counts = result.get_counts()
plot_histogram(counts)
```

$\mbox{Estado inicial} = \frac{1}{\sqrt{2}}(|0\rangle + i|1\rangle)$

```python
# Definimos el estado |q_0>
initial_state = [1/sqrt(2), 1j/sqrt(2)]  
# Redefinimos el circuito cuantico
qc = QuantumCircuit(1) 
# Inicializamos el qubit 0 en el estado inicial "initial_state"
qc.initialize(initial_state, 0) 
# Almacenamos el vector de estado
qc.save_statevector() 
qobj = assemble(qc)
# Ejecutamos el circuito cuantico
state = sim.run(qobj).result().get_statevector() 
# Imprimimos el resultado 
print(state)          
```

```python
qobj = assemble(qc)
results = sim.run(qobj).result().get_counts()
plot_histogram(results)
```

```python
qc = QuantumCircuit(1) # We are redefining qc
initial_state = [0.+1.j/sqrt(2),1/sqrt(2)+0.j]
qc.initialize(initial_state, 0)
qc.draw("mpl")
```

```python
qc.save_statevector()
result = sim.run(assemble(qc)).result()
state = result.get_statevector()
print("Qubit State = " + str(state))
```
